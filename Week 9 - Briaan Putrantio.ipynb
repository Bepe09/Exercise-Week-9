# Association Rules

Association Rule Mining adalah salah satu metode data mining yang dapat mengidentifikasi hubungan kesamaan antar item. Dalam metode ini dibutuhkan algoritma untuk mencari kandidat aturan asosiasi.

Association Rules merupakan "if-then" statements, yang membantu menunjukkan kemungkinan hubungan antara item data, dalam kumpulan data besar di berbagai jenis database. Association Rule Mining memiliki sejumlah cara pengaplikasian dan digunakan secara luas untuk membantu menemukan korelasi penjualan dalam data transaksional atau dalam kumpulan data medis.

# Tujuan Association Rules Mining

Association Rules Mining bertujuan untuk menunjukkan nilai asosiatif antara jenis-jenis produk yang dibeli oleh pelanggan sehingga terlihatlah suatu pola berupa produk apa saja yang sering dibeli oleh pelanggan tersebut. Dengan mengetahui jenis produk yang sering dibeli maka dapat dibuat sebagai sebuah dasar keputusan untuk menentukan produk apa saja yang cocok untuk dipromosikan kepada pelanggan tersebut. Algoritma Apriori juga perlu dipergunakan untuk menentukan frequent itemset sehingga hasil akhir yang dicapai yaitu untuk menghitung persentase ketertarikan (confindence) pelanggan terhadap produk yang ditawarkan.

Terdapat 2 bagian dalam Association Rules :

- Antecedent (if)
- Consequent (then)

Antecedent merupakan sesuatu yang ditemukan di dalam data, dan Consequent adalah item yang ditemukan di dalam hubungannya dengan Antecedent. 

Contoh nya seperti aturan berikut:
"70% dari orang-orang yang membeli mie, juice dan saus akan membeli juga roti tawar".

# Association Rules Mining : Metrik Evaluasi

<b>Metrik evaluasi dalam Association Rule Mining adalah sebagai berikut :

- Support (s) => Ini adalah jumlah transaksi yang menyertakan item dari {X} dan {Y} bagian aturan sebagai persentase dari total transaksi. Ini dapat direpresentasikan dalam bentuk persentase dari semua transaksi yang menunjukkan seberapa sering sekelompok item terjadi bersama-sama.

Support = σ(X+Y) ÷ total => Ini adalah pecahan dari transaksi yang mencakup X dan Y.


- Confidence(c) => Rasio ini menunjukkan jumlah total transaksi semua item di {A} dan {B} dengan jumlah transaksi item di {A}.

Conf(X=>Y) = Supp(X∪Y) Supp(X) => Menghitung berapa kali setiap item di Y muncul dalam transaksi yang juga menyertakan item di X.


- Lift(l) => Kenaikan aturan X=>Y adalah kepercayaan aturan dibagi dengan kepercayaan yang diharapkan. di sini, diasumsikan bahwa itemset X dan Y independen satu sama lain. Keyakinan yang diharapkan dihitung dengan membagi keyakinan dengan frekuensi {Y}.

Lift(X=>Y) = Conf(X=>Y) Supp(Y) => Nilai lift yang mendekati 1 menunjukkan bahwa X dan Y hampir selalu muncul bersamaan seperti yang diharapkan. Nilai peningkatan lebih besar dari 1 menunjukkan bahwa mereka muncul bersama lebih dari yang diharapkan, dan nilai peningkatan kurang dari 1 menunjukkan bahwa mereka muncul kurang dari yang diharapkan. Nilai peningkatan yang lebih besar menunjukkan asosiasi yang lebih kuat.

# Implementasi Association Rule Mining

### *implementasi menggunakan apriori*

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Bepe09/Exercise-Week-9/main/Retail%20Online.csv", header=None, sep=",")
data.head(10)
```

```python
data = data.fillna(" ")
data.head(10)
```

```python
list_data = set()
for i in data:
    if data[i].unique != " " :
        list_data.update(data[i].unique())
print(list_data)
```

```python
set_list = set(list_data)
encoded = []
for i, j in data.iterrows():
    rowset = set(j) 
    labels = {}
    uncommons = list(set_list - rowset)
    commons = list(set_list.intersection(rowset))
    for k in uncommons:
        labels[k] = 0
    for z in commons:
        labels[z] = 1
    encoded.append(labels)
encoded[0]
df = pd.DataFrame(encoded)
```

```python
df_banyak = apriori(df, min_support=0.2, use_colnames=True, verbose=1)
df_banyak
```

```python
dr_ar = association_rules(df_banyak, metric="confidence", min_threshold=0.6)
dr_ar.head()
```
